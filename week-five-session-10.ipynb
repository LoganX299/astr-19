```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
random_numbers = np.random.wald(5, 25, 1000)

plt.hist(random_numbers, bins=100, color='purple', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Numbers Randomly Distributed by Wald")

plt.savefig("Wald_Histogram.pdf")

plt.show()
```
