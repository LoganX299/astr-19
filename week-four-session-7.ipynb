```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1 ,100)

def exponential(x):
    return np.exp(x)
```

```python
y = exponential(x)

plt.plot(x, y)
plt.xlabel("Time [Milliseconds]")
plt.ylabel("Awesomeness")
plt.title("The Exponential Growth of Awesomeness Over Time in Milliseconds")

plt.savefig("exp_plot.pdf")

plt.show()
```
